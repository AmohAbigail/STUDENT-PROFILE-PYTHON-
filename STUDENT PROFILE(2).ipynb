```python
### print("==== STUDENT PROFILE SYSTEM ====\n")


name = input('Enter your full name: ')
age = input(input('Enter your age: '))
gpa = float(input('Enter your GPA (0.0 - 4.0'))
courses_input =input('Enter your courses (comma-seperated): ')
courses = [course.strip() for course in courses_input.split(',')]

student_profile ={
    "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": len(courses)
}


print("\n" + "="*50)
print("STUDENT PROFILE")
print("="*50)
print(f"Full Name: {student_profile['name']}")
print(f"Age: {student_profile['age']}")
print(f"GPA: {student_profile['gpa']:.2f}")
print(f"Numer of courses: {student_profile['num_courses']}")
print("Courses Enrolled:")
for i, course in enumerate(student_profile['courses'], 1):
    print(f" {1}. {course}")
print("="*50)

print("\n--- TYPE CONVERSION DEMESTRATION -----")
print(f"Name type: {type(student_profile['name'])} -> {type(str(student_profile['name']))}")
print(f"Age type: {type(student_profile['age'])} -> converted from string to int")
print(f"GPA type: {type(student_profile['gpa'])} -> converted from string to float")
print(f"Courses type: {type(student_profile['courses'])} -> List from comma-separated string")
print(f"Num Courses type: {type(student_profile['num_courses'])} -> Calculated from list length")
print("\n=== Student Profile completed successfully!")
```
